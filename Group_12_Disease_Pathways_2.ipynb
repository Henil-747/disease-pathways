# Generating Graphs


## Complete PPI Network Plot


```python
from pyvis.network import Network
import pandas as pd
import numpy as np
import networkx as nx

got_net = Network(height='750px', width='100%', bgcolor='#222222', font_color='white')

# set the physics layout of the network
got_net.barnes_hut()
got_data = pd.read_csv('/content/bio-pathways-network.csv')

sources = got_data['Gene ID 1']
targets = got_data['Gene ID 2']

edge_data = zip(sources, targets)

for e in edge_data:
    src = e[0]
    dst = e[1]

    got_net.add_node(src, src, title=src)
    got_net.add_node(dst, dst, title=dst)
    got_net.add_edge(src, dst)

neighbor_map = got_net.get_adj_list()

# add neighbor data to node hover data
for node in got_net.nodes:
    node['title'] += ' Neighbors:<br>' + '<br>'.join(neighbor_map[node['id']])
    node['value'] = len(neighbor_map[node['id']])

got_net.show('Whole Protein-Protein Interaction.html')
```

##Generatng graph for Inherited Metabolic Disorder Class

```python
df = pd.read_csv('/content/bio-pathways-diseaseclasses.csv')
df1 = df[df['Disease Class'] == "inherited metabolic disorder"]

arr1 = []
for id in df1['Disease ID']:
  arr1.append(id)

arr2 = []
df2 = pd.read_csv('/content/bio-pathways-associations.csv')
for id in arr1:
  df3 = df2[df2['Disease ID'] == id]
  temp1 = df3['Associated Gene IDs']
  temp2 = temp1.to_numpy()
  temp3 = temp2[0].split(", ")
  arr2 += temp3

test1 = list(set(arr2))
print(test1)
print(len(test1))

arr3 = []
G = nx.Graph()
got_net = Network(height='750px', width='100%', bgcolor='#222222', font_color='white')

df4 = pd.read_csv('/content/bio-pathways-network.csv')
for i in range(0, len(test1)):

    oop = df4[df4['Gene ID 1'] == int(test1[i])]
    oop2 = df4[df4['Gene ID 2'] == int(test1[i])]

    for k in oop['Gene ID 2']:
      if str(k) in test1:
        got_net.add_node(str(test1[i]), str(test1[i]), title=str(test1[i]), color="#ffff00")
        got_net.add_node(str(k),str(k),title=str(k), color="#ffff00")
        got_net.add_edge(str(test1[i]), str(k))
        G.add_edge(str(test1[i]), str(k))
        
    for l in oop2['Gene ID 1']:
      if str(l) in test1:
        got_net.add_node(str(test1[i]), str(test1[i]), title=str(test1[i]), color="#ffff00")
        got_net.add_node(str(l),str(l),title=str(l),color="#ffff00")
        got_net.add_edge(str(l), str(test1[i]))
        G.add_edge(str(l), str(test1[i]))

neighbor_map = got_net.get_adj_list()

for node in got_net.nodes:
    node['title'] += ' Neighbors:<br>' + '<br>'.join(neighbor_map[node['id']])
    node['value'] = len(neighbor_map[node['id']])

for i in test1:
  if i not in G.nodes():
    got_net.add_node(str(i), str(i), title=str(i), color="#ffffff") 


got_net.show('inherited_metabolic_disorder.html')
```

```python
# For Inherited Metabolic Disorder (Disease Class)

shortest_path = dict(nx.shortest_path_length(G))
Hemochromatosis_associated_gene = [1356, 148738, 30061, 3077, 3240, 4891, 57817, 650, 654, 6718, 7036, 7124]
LeighDisease_associated_gene = [123263, 1355, 137682, 1738, 1892, 374291, 4508, 4514, 4535, 4536, 4537, 4538, 4540, 4541, 4566, 4567, 4577, 4578, 4695, 4720, 4722, 4723, 4724, 4728, 51204, 5428, 55572, 55699, 617, 6389, 6834, 79133, 79731]
Hyperhomocysteinemia_associated_gene = [113235, 1674, 27232, 29108, 4524, 4552, 4868, 5444, 7124, 834, 875]
combined_associated_gene = Hemochromatosis_associated_gene + LeighDisease_associated_gene + Hyperhomocysteinemia_associated_gene

# print(combined_associated_gene)
distance_individual_all_three = []
# for count in range(2):
for i in range(len(combined_associated_gene)):
  for j in range(len(combined_associated_gene)):
   try:
     distance_individual_all_three.append(shortest_path[str(combined_associated_gene[i])][str(combined_associated_gene[j])])
   except:
     pass

avg_spread_among_three_disease = (sum(distance_individual_all_three))/(len(distance_individual_all_three))
print(f'Avg spread for three disease among the whole network: {avg_spread_among_three_disease}')

#############################################################################
distance_individual_Hemochromatosis_Hyperhomocysteinemia = []
for k in range(len(Hemochromatosis_associated_gene)):
  for l in range(len(Hyperhomocysteinemia_associated_gene)):
   try:
     distance_individual_Hemochromatosis_Hyperhomocysteinemia.append(shortest_path[str(Hemochromatosis_associated_gene[k])][str(Hyperhomocysteinemia_associated_gene[l])])
   except:
     pass

avg_spread_among_Hemochromatosis_Hyperhomocysteinemia = (sum(distance_individual_Hemochromatosis_Hyperhomocysteinemia))/(len(distance_individual_Hemochromatosis_Hyperhomocysteinemia))
print(f'Avg spread among Hemochromatosis_Hyperhomocysteinemia : {avg_spread_among_Hemochromatosis_Hyperhomocysteinemia}')

############################################################################
distance_individual_Hemochromatosis_LeighDisease = []
for k in range(len(Hemochromatosis_associated_gene)):
  for l in range(len(LeighDisease_associated_gene)):
   try:
     distance_individual_Hemochromatosis_LeighDisease.append(shortest_path[str(Hemochromatosis_associated_gene[k])][str(LeighDisease_associated_gene[l])])
   except:
     pass

avg_spread_among_Hemochromatosis_LeighDisease = (sum(distance_individual_Hemochromatosis_LeighDisease))/(len(distance_individual_Hemochromatosis_LeighDisease))
print(f'Avg spread among Hemochromatosis_LeighDisease : {avg_spread_among_Hemochromatosis_LeighDisease}')
#############################################################################

distance_individual_Hyperhomocysteinemia_LeighDisease = []
for k in range(len(Hyperhomocysteinemia_associated_gene)):
  for l in range(len(LeighDisease_associated_gene)):
   try:
     distance_individual_Hyperhomocysteinemia_LeighDisease.append(shortest_path[str(Hyperhomocysteinemia_associated_gene[k])][str(LeighDisease_associated_gene[l])])
   except:
     pass

avg_spread_among_Hyperhomocysteinemia_LeighDisease = (sum(distance_individual_Hyperhomocysteinemia_LeighDisease))/(len(distance_individual_Hyperhomocysteinemia_LeighDisease))
print(f'Avg spread among Hyperhomocysteinemia_LeighDisease : {avg_spread_among_Hyperhomocysteinemia_LeighDisease}')

```
